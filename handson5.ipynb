Problem Statement 5: Temperature Converter
Objective:
Create a Python program that converts temperatures between Celsius, Fahrenheit, and Kelvin using functions.

Background:
Temperature conversion is a common task in science and everyday life. In this problem, you will create a program that allows users to convert temperatures from one scale to another by defining and using functions. This will help you understand how to encapsulate logic within functions and reuse code.

Requirements:

Input:

The program should prompt the user to enter a temperature value.
The user should also specify the current scale of the temperature (Celsius, Fahrenheit, or Kelvin) and the scale to which they want to convert.
Process:

Create separate functions for each type of conversion:
celsius_to_fahrenheit(celsius)
celsius_to_kelvin(celsius)
fahrenheit_to_celsius(fahrenheit)
fahrenheit_to_kelvin(fahrenheit)
kelvin_to_celsius(kelvin)
kelvin_to_fahrenheit(kelvin)
Use these functions to perform the conversion based on user input.
Output:

Display the converted temperature value with the appropriate scale.
Bonus Features:

Allow the user to convert multiple temperatures in one session.
Provide a summary of all conversions performed in the session.

```python
def temperature_converter(value, current_scale, desired_scale):
    if current_scale == "celsius" and desired_scale == "fahrenheit": 
        desired_value = celsius_to_fahrenheit(value)
        return print(f"Converted temperature is {desired_value} F")
    
    if current_scale == "fahrenheit" and desired_scale == "celsius": 
        desired_value = fahrenheit_to_celsius(value)
        return print(f"Converted temperature is {desired_value} C")
    
    if current_scale == "celsius" and desired_scale == "kelvin": 
        desired_value = celsius_to_kelvin(value)
        return print(f"Converted temperature is {desired_value} K")

    if current_scale == "kelvin" and desired_scale == "celsius": 
        desired_value = kelvin_to_celsius(value)
        return print(f"Converted temperature is {desired_value} C")

    if current_scale == "kelvin" and desired_scale == "fahrenheit": 
        desired_value = kelvin_to_fahrenheit(value)
        return print(f"Converted temperature is {desired_value} F")

    if current_scale == "fahrenheit" and desired_scale == "kelvin": 
        desired_value = fahrenheit_to_kelvin(value)
        return print(f"Converted temperature is {desired_value} K")

    else :
        print("please enter a correct scale value")
        return


    

def celsius_to_fahrenheit(value):
    desired_value = 1.8 * value + 32 
    return desired_value

def celsius_to_kelvin(value):
    desired_value = value + 273 
    return desired_value

def kelvin_to_celsius(value):
    desired_value = value - 273 
    return desired_value

def kelvin_to_fahrenheit(value):
    desired_value = (value - 273) * 1.8 + 32 
    return desired_value

def fahrenheit_to_kelvin(value):
    desired_value = (value - 32) * 5/9 +273 
    return desired_value

def fahrenheit_to_celsius(value):
    desired_value = (value - 32) * 5/9 
    return desired_value

  

    
    
```

```python
value = float(input("Please enter the temp value"))
current_scale = input("Please enter the current scale (celsius, Fahrenheit, or Kelvin)")
current_scale = current_scale.lower()
desired_scale = input("Please enter the desired scale (celsius, Fahrenheit, or Kelvin)")
desired_scale = desired_scale.lower()

temperature_converter(value, current_scale, desired_scale)
```
